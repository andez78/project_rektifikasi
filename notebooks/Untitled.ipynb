```python
# Import library yang dibutuhkan
import matplotlib.pyplot as plt

# --- Data Contoh (Ganti dengan data Anda jika ada) ---
penyebab = ['Menunggu Material AOG', 'Keterbatasan Personel/Shift', 'Ketersediaan Tools & Equipment', 'Lain-lain']
persentase = [55, 20, 15, 10]
colors = ['#F44336', '#FFC107', '#2196F3', '#9E9E9E'] # Warna-warna yang kontras
# 'explode' untuk menonjolkan irisan pertama (Menunggu Material Ad-hoc)
explode = (0.05, 0, 0, 0)  

# --- Membuat Visualisasi Diagram Donat ---
fig, ax = plt.subplots(figsize=(8, 8))

# Membuat pie chart
wedges, texts, autotexts = ax.pie(persentase,
                                  explode=explode,
                                  labels=penyebab,
                                  colors=colors,
                                  autopct='%1.1f%%', # Format angka persentase
                                  shadow=True,
                                  startangle=90, # Atur sudut awal
                                  pctdistance=0.85, # Jarak persentase dari pusat
                                  textprops=dict(color="w", fontsize=10, fontweight='bold')) # Styling teks persentase

# Menambahkan judul
ax.set_title('Faktor Utama Penyebab Keterlambatan Rektifikasi Order', fontsize=14, pad=20)

# Membuat lingkaran putih di tengah untuk efek "Donat"
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

# Memastikan rasio aspek sama agar lingkaran terlihat bulat sempurna
ax.axis('equal')  

# Menambahkan legenda
ax.legend(wedges, penyebab,
          title="Penyebab Keterlambatan",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

# Atur agar teks label tidak tumpang tindih dengan teks persentase
plt.setp(autotexts, size=12, weight="bold")
plt.setp(texts, size=0) # Sembunyikan label di samping irisan, karena sudah ada di legenda

# Menampilkan diagram
plt.tight_layout() # Atur layout agar pas
plt.show()

# --- Untuk menyimpan diagram sebagai file gambar (PNG) ---
# Hapus tanda pagar (#) pada baris di bawah ini untuk menyimpan gambar
fig.savefig('diagram_donat_penyebab_keterlambatan.png', dpi=300, bbox_inches='tight')
print("Diagram donat berhasil disimpan sebagai 'diagram_donat_penyebab_keterlambatan.png'")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# --- Membuat Visualisasi Gunung Es ---
fig, ax = plt.subplots(figsize=(12, 9)) # Ukuran gambar (lebar, tinggi)

# Sembunyikan sumbu X dan Y
ax.set_xlim(0, 12)
ax.set_ylim(0, 9)
ax.axis('off')

# --- Menggambar Elemen Visual ---
# Latar belakang biru untuk laut dan langit
ax.axhspan(0, 4.5, color='#E3F2FD', zorder=0) # Laut
ax.axhspan(4.5, 9, color='#F0F8FF', zorder=0) # Langit

# Garis permukaan air
ax.plot([0, 12], [4.5, 4.5], color='#0D47A1', linewidth=2)
ax.text(0.2, 4.6, 'Permukaan Air', fontsize=10, color='#0D47A1')

# Menggambar gunung es
# Bagian atas (terlihat)
iceberg_top_points = np.array([[5, 4.5], [6, 6.5], [7, 4.5]])
iceberg_top = patches.Polygon(iceberg_top_points, closed=True, facecolor='white', edgecolor='#B0E0E6', linewidth=2)
ax.add_patch(iceberg_top)

# Bagian bawah (tersembunyi)
iceberg_bottom_points = np.array([[5, 4.5], [7, 4.5], [10, 1], [2, 1]])
iceberg_bottom = patches.Polygon(iceberg_bottom_points, closed=True, facecolor='#B0E0E6', edgecolor='#4682B4', linewidth=2)
ax.add_patch(iceberg_bottom)

# --- Menambahkan Teks ---

# 1. Teks di Puncak Gunung Es (Masalah Inti)
ax.text(6, 5.8, 'Masalah Inti (Gejala)', ha='center', va='center', fontsize=14, fontweight='bold', color='#BF360C')
ax.text(6, 5.1,
        'Proses rektifikasi tidak terstandar,\nmenyebabkan inkonsistensi prosedur\nserta waktu penyelesaian yang lambat.',
        ha='center', va='top', fontsize=11, wrap=True)

# 2. Teks di Bawah Air (Akar Penyebab)
ax.text(6, 3.8, 'Akar Penyebab (Masalah Sebenarnya)', ha='center', va='center', fontsize=14, fontweight='bold', color='#0D47A1')
ax.text(6, 3.2,
        'Data historis diperlakukan sebagai arsip pasif.\nHal ini menciptakan ketergantungan pada pengalaman individu\ndan menyebabkan knowledge bottleneck.',
        ha='center', va='top', fontsize=11, wrap=True)

# 3. Teks di Dasar (Dampak Bisnis)
ax.text(6, 2.0, 'Dampak Bisnis (Konsekuensi Terbesar)', ha='center', va='center', fontsize=14, fontweight='bold', color='#0D47A1')
ax.text(6, 1.4,
        ' Dampak Finansial: Jam kerja tidak produktif, downtime mahal, biaya rework.\n Dampak SDM: Risiko knowledge drain & lambatnya pengembangan kompetensi.',
        ha='center', va='top', fontsize=11, wrap=True)

# Judul Utama
fig.suptitle('Problem Statement: Analogi Gunung Es', fontsize=20, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# --- Untuk menyimpan diagram sebagai file gambar (PNG) ---
# Hapus tanda pagar (#) pada baris di bawah ini untuk menyimpan gambar
# fig.savefig('diagram_gunung_es.png', dpi=300, bbox_inches='tight')
# print("Diagram berhasil disimpan sebagai 'diagram_gunung_es.png'")
```

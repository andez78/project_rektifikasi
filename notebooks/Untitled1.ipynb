```python
# Import library yang dibutuhkan
import matplotlib.pyplot as plt
import numpy as np

# --- Data Contoh (Ganti dengan persentase atau nilai absolut Anda) ---
aktivitas = ['Total Waktu Rektifikasi']
waktu_produktif = [65]  # Persentase untuk eksekusi
waktu_pencarian_info = [25] # Persentase untuk mencari info
waktu_koordinasi = [10] # Persentase untuk koordinasi & konsultasi

# --- Membuat Visualisasi Diagram Batang Bertumpuk Horizontal ---
fig, ax = plt.subplots(figsize=(10, 4)) # Ukuran gambar (lebar, tinggi)

# Membuat bar secara bertumpuk
ax.barh(aktivitas, waktu_produktif, label='Eksekusi RO (Produktif)', color='#4CAF50')
ax.barh(aktivitas, waktu_pencarian_info, left=waktu_produktif, label='Pencarian Referensi (AMM/SRM)', color='#FFC107')
ax.barh(aktivitas, waktu_koordinasi, left=np.add(waktu_produktif, waktu_pencarian_info), label='Konsultasi & Koordinasi', color='#F44336')

# --- Styling untuk Tampilan Profesional ---
ax.set_xlabel('Persentase Alokasi Waktu (%)', fontsize=12)
ax.set_title('Komposisi Waktu dalam Proses Rektifikasi Order', fontsize=14, pad=20)
ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.4), ncol=3, frameon=False, fontsize=11)

# Menghilangkan bingkai yang tidak perlu
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_yticks([]) # Sembunyikan label sumbu Y karena sudah jelas

# Menambahkan teks persentase di dalam setiap segmen bar
total = waktu_produktif[0] + waktu_pencarian_info[0] + waktu_koordinasi[0]
p1_x = waktu_produktif[0] / 2 # Posisi teks di tengah segmen pertama
p2_x = waktu_produktif[0] + waktu_pencarian_info[0] / 2 # Posisi teks di tengah segmen kedua
p3_x = waktu_produktif[0] + waktu_pencarian_info[0] + waktu_koordinasi[0] / 2 # Posisi teks di tengah segmen ketiga

ax.text(p1_x, 0, f'{waktu_produktif[0]}%', ha='center', va='center', color='white', fontsize=12, fontweight='bold')
ax.text(p2_x, 0, f'{waktu_pencarian_info[0]}%', ha='center', va='center', color='black', fontsize=12, fontweight='bold')
ax.text(p3_x, 0, f'{waktu_koordinasi[0]}%', ha='center', va='center', color='white', fontsize=12, fontweight='bold')

# Atur batas sumbu X agar pas
ax.set_xlim(0, total)

# Menampilkan diagram
plt.tight_layout()
plt.show()

# --- Untuk menyimpan diagram sebagai file gambar (PNG) ---
# Hapus tanda pagar (#) pada baris di bawah ini untuk menyimpan gambar
# fig.savefig('diagram_waktu_produktif.png', dpi=300, bbox_inches='tight')
# print("Diagram berhasil disimpan sebagai 'diagram_waktu_produktif.png'")
```
